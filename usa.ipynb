```python
dataset="./us-accidents/usaa.csv" 
```

```python
import pandas as pd 
```

```python
df=pd.read_csv(dataset) 
df
```

```python
df.columns
```

```python
df.describe()
```

```python
#handel missing value 
missing=df.isnull().sum().sort_values(ascending=False)/len(df) 
missing
```

```python
#drop 0 values 
missing[missing !=0].plot(kind='barh')
```

```python
# data visulizataion  
```

```python
df.City.value_counts().unique()
```

```python
city_value=df.City.value_counts() 
city_value
```

```python
city_value[:10].plot(kind="barh")
```

```python
import seaborn as sns
sns.set_theme(style='darkgrid')
```

```python
sns.distplot(city_value)
```

```python
#accident more then 1000 
sns.distplot(city_value >=1000)
```

```python
# accident value lessthen 1000 
sns.distplot(city_value<1000)
```

```python
#accident more then 500 on the city 
temp=city_value[city_value>500] 

sns.histplot(temp,log_scale=True)
```

```python
df.columns
```

```python
df.Start_Time
```

```python
#to convort data time 
date_time=pd.to_datetime(df.Start_Time)
```

```python
date_time[0]
```

```python
#get hours from date time 
sns.distplot(date_time.dt.hour,bins=24)
```

```python
# day  
sns.histplot(date_time.dt.dayofweek)
```

```python
#find which time more accident  
sunday=date_time[date_time.dt.dayofweek == 6] 

sns.distplot(sunday.dt.hour,norm_hist=True,bins=24)
```

```python
#find the acciden in which month 
sns.distplot(date_time.dt.month,norm_hist=True,bins=12)
```

```python
#2018 usa accident data 
data_2018=date_time[date_time.dt.year == 2018] 
sns.distplot(data_2018.dt.month,norm_hist=True,bins=12)
# data_2018
```
